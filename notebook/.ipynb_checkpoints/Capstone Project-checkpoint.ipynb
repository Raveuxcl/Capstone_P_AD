```python
df_llu=pd.read_excel("C:/Users/galla/OneDrive/Documentos/GitHub/Capstone_P_AD-1/data/processed/Consolidado_lluvia.xlsx")
df_llu.head(10)
```

### Sin embargo, estos datos no vienen limpios, por lo que es necesario trabajarlo. Se procede a cambiar los datos nulos 's/p' por 0 para que se puedan trabajar de forma numerica.

### Como podemos ver, el tipo de dato no nos sirve para poder desarrollar el analisis, por lo que es necesario cambiar el tipo a numerico(float).

```python
df_llu['E. El Paico']=pd.to_numeric(df_llu['E. El Paico'],errors='coerce')
df_llu['E. Eulogio Sanchez']=pd.to_numeric(df_llu['E. Eulogio Sanchez'],errors='coerce')
df_llu['E. Pudahuel']=pd.to_numeric(df_llu['E. Pudahuel'],errors='coerce')
df_llu['E. Quinta Normal']=pd.to_numeric(df_llu['E. Quinta Normal'],errors='coerce')
df_llu['E. San Jose']=pd.to_numeric(df_llu['E. San Jose'],errors='coerce')
```

```python
df_llu.info()
```

### Luego de forzar el cambio a numero decimal, podemos ver que el cambio se hizo efectivo y todas las variables son de tipo float, excepto la variable de fecha.
### Por otro lado, podemos que que para la comuna San Jose, existen menos valores, por lo que es necesario completar la esta comuna con ceros.

```python
df_llu=df_llu.fillna(0)
```

```python
df_llu.info()
```

### Ahora cargaremos la base de servicio movil UPTIME, la cual fue obtenida de manera privada, que nos muestra el estado del servicio en el gran Santiago.

```python
df_upt=pd.read_csv ('C:/Users/galla/OneDrive/Documentos/GitHub/Capstone_P_AD-1/data/raw/Raw-Uptime/raw.uptime.20222.csv',sep=";",decimal=',')
df_upt.head(10)
```

```python
df_upt = df_upt.set_index(pd.to_datetime(df_upt.fecha, format='%d-%m-%Y'))
df_upt.head(10)
```

### Luego procedemos a unir con las bases cargadas y trabajadas anteriormente, df_llu y df_temp.

```python
df_upt['temp_media']= df_upt.fecha.map(df_temp.set_index(df_temp.Fecha)['promedio_estaciones'])
df_upt['lluvia']= df_upt.fecha.map(df_llu.set_index(df_llu.Fecha)['promedio_estaciones'])
df_upt=df_upt.fillna(0)
```

```python
#df_upt['upt_3g']=pd.to_numeric(df_upt['upt_3g'],errors='coerce')
df_upt.info()
```

```python
df_upt.head()
```

### Como se puede ver, se unieron las bases correctamente.

```python
#df_upt = df_upt.set_index(pd.to_datetime(df_upt.fecha, format='%Y-%m-%d'))
#df_upt=df_upt.set_index('fecha')
#df_upt.plot()

fig, ax = plt.subplots(figsize=(15,6))

ax.plot(df_upt["fecha"], df_upt["upt_3g"], label ="UPTIME")
ax.plot(df_upt["fecha"], df_upt["temp_media"], label ="Temperatura Media")
ax.plot(df_upt["fecha"], df_upt["lluvia"], label ="Lluvia Media(mm)")
#ax.scatter(df_llu["Fecha"], df_llu["E. Eulogio Sanchez"])
#ax.scatter(df_llu["Fecha"], df_llu["E. Pudahuel"])
#ax.scatter(df_llu["Fecha"], df_llu["E. Quinta Normal"])
#ax.scatter(df_llu["Fecha"], df_llu["E. San Jose"])
ax.legend()
ax.set_title("Temperatura Promedio por dia")

fig.tight_layout()
```

### A priori podemos ver que cuando existen peaks de temperatura, baja el servicio uptime, al igual que cuando llueve una gran cantidad.

### Ahora se ordenara el data set por dia, para que sea dividida en dos subsets, Train y Test, esto para entrenar el modelo y probarlo.

```python
df_upt= df_upt.sort_index().asfreq(freq='D')
df_upt
```

```python
(df_upt.index == pd.date_range(
                    start = df_upt.index.min(),
                    end   = df_upt.index.max(),
                    freq  = df_upt.index.freq)
).all()
```

### Luego procedemos a dividir la base en dos, df_upt_train y df_upt_test.
 

```python
steps = 10
df_upt_train = df_upt[:-steps]
df_upt_test  = df_upt[-steps:]

print(f"Fechas train : {df_upt_train.index.min()} --- {df_upt_train.index.max()}  (n={len(df_upt_train)})")
print(f"Fechas test  : {df_upt_test.index.min()} --- {df_upt_test.index.max()}  (n={len(df_upt_test)})")

fig, ax = plt.subplots(figsize=(15, 8))
df_upt_train['upt_3g'].plot(ax=ax, label='train')
df_upt_test['upt_3g'].plot(ax=ax, label='test')
ax.legend();
```
